```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use("dstyle")
_ = plt.figure()
plt.style.use("dstyle")
_ = plt.figure()
```

## Sod Test

```python
soln_fname = "sod_soln.txt"

soln_info = np.loadtxt(soln_fname, delimiter=',')
soln_head = soln_info[0]
data = soln_info[2:]
domlen, ncells, diaph, timeout, _ = soln_head
dx = domlen/ncells
ncells = int(ncells)

DL, UL, PL = 1.0, 0.0, 1.0
DR, UR, PR = 0.125, 0.0, 0.1
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(12,5))

x_arr = (np.arange(0,int(ncells))+0.5)*dx
ind_stateL = diaph > x_arr
ind_stateR = diaph <= x_arr

density0 = np.zeros(ncells)
density0[ind_stateL] = DL
density0[ind_stateR] = DR

velocity0 = np.zeros(ncells)
velocity0[ind_stateL] = UL
velocity0[ind_stateR] = UR

pressure0 = np.zeros(ncells)
pressure0[ind_stateL] = PL
pressure0[ind_stateR] = PR

ax[0].plot(x_arr, density0)
ax[0].set_title("Density")

ax[1].plot(x_arr, velocity0)
ax[1].set_title("Velocity")

ax[2].plot(x_arr, pressure0)
ax[2].set_title("Pressure")

fig.suptitle(f"Time = 0.00")
fig.tight_layout()
fig.subplots_adjust(top=0.75)
```

```python
xposns = data[:,0]
densities = data[:,1]
velocities = data[:,2]
pressures = data[:,3]
energies = data[:,4]
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

ax[0,0].plot(xposns, densities)
ax[0,0].set_ylabel("Density")
ax[0,0].set_xlim(0,1), ax[0,0].set_ylim(0,1.2)

ax[0,1].plot(xposns, velocities)
ax[0,1].set_ylabel("Velocity")
ax[0,1].set_xlim(0,1), ax[0,1].set_ylim(0, 1.2)

ax[1,0].plot(xposns, pressures)
ax[1,0].set_ylabel("Pressure")
ax[1,0].set_xlim(0,1), ax[1,0].set_ylim(0,1.2)

ax[1,1].plot(xposns, energies)
ax[1,1].set_ylabel("Internal Energy")
ax[1,1].set_xlim(0,1)

fig.suptitle(f"Time = {timeout:.3f}")
fig.tight_layout()
fig.subplots_adjust(top=0.85)
```
